```python
#Cargar la extension de memory_profiler
%load_ext memory_profiler
```

```python
#Preparar perfilador de python profiler
pr = cProfile.Profile()
```

```python
# q1t
pr.enable()

resultq1t = q1t.q1_time(file_path)

pr.disable()

# Primero: se crea un objeto StringIO para capturar la salida del análisis de rendimiento
# Segundo: se crea un objeto Stats y usa print_stats para imprimir el informe
s = io.StringIO()
ps = pstats.Stats(pr, stream=s).sort_stats('cumulative')
ps.print_stats()
```

```python

print(s.getvalue())
```
